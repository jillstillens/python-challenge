```python
### Import the pathlib and csv library
from pathlib import Path
import csv

### Set File paths
#Input csv file: contains two columns: date and profit-and-loss (pnl) amount
csvpath = Path('Resources/budget_data.csv')
#Output txt file: will contain summary statistics
output = Path("output.txt")


### Initialize variables
records = {}
total=0
nbr_months=0
min_dt = ""
min_pnl = 99999999
max_dt = ""
max_pnl = 0
prev_pnl = 0
pnl_diff_total = 0

### Functions used in this program
# Print record to both console and file
def print_it(toPrint):
    print(toPrint)
    print(toPrint,file=f)
    return

### Start of Main Logic
# Open the csv file as an object
with open(csvpath, 'r') as csvfile:

    # Pass in the csv file to the csv.reader() function
    # (with ',' as the delmiter/separator) and return the csvreader object
    csvreader = csv.reader(csvfile, delimiter=',')

    # Read the header row first, to get past it and into the data
    csv_header = next(csvreader)
    
    #Read and load entire file into dictionary "records"
    records = {rows[0]: int(rows[1]) for rows in csvreader}

# Cycle through each key\value (date\profit-and-loss amt) pair in the dictionary 
for key,value in records.items():
    #set smallest profit and loss amt on file
    if value < min_pnl:
        min_pnl = value
        min_dt = key
    #set max profit and loss amt on file
    if value > max_pnl:
        max_pnl = value
        max_dt = key
    #total pnl differences between each month
    # first time through, don't calculate difference between months since there is no 
    # previous value; therefore just set previous value
    if prev_pnl == 0:
        prev_pnl = value
    else:
        # if not first time through, keep running total of difference between current and prev month
        pnl_diff_total = pnl_diff_total + (value - prev_pnl)
        prev_pnl = value

### Perform overall calculations
#set total number of months based on length of dictionary:
nbr_months = len(records)
#sum all pnl values
total=sum(records.values())
#calc average change between months; note have to skip first months since
#no previous value, so subtract one from total number of months
avg_change = pnl_diff_total / (nbr_months - 1)

### Write formatted output; note dollars are formatted to append "$" with commas and decimals where appropriate
with open(output, 'w') as f:
    record = "Financial Analysis"
    print_it(record)
    record = "------------------"
    print_it(record)
    record = "Total Months: " + str(nbrMonths)
    print_it(record)
    record = "Total: " + "${:,.0f}".format(total)
    print_it(record)
    record = "Average Change: " + " ${:,.2f}".format(avg_change)
    print_it(record)
    record = "Greatest increase in profits: " + max_dt + " ${:,.0f}".format(max_pnl)
    print_it(record)
    record = "Greatest decrease in profits: " + min_dt + " ${:,.0f}".format(min_pnl)
    print_it(record)

### End of program
```
